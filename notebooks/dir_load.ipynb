```
./iocamlserver.byte -static /home/andyman/.ipython/profile_iocamljs -serve test test
```

This enables iocamljs and serves the directory test which should have a file `test/test.cmo`

The following function loads a file from the server and registers it with js_of_ocaml

````ocaml
let register_file fname = 
    match Iocaml.load_from_server fname with
    | Some(x) -> Sys_js.register_file ~name:fname ~content:x
    | None -> ()
````
