# An attempt to #load in js_of_ocaml

This should be run with something like

```
./iocamlserver.byte -static /home/andyman/.ipython/profile_iocamljs -serve test test
```

This enables iocamljs and serves the directory test which should have a file `test/test.cmo`

The following function loads a file from the server and registers it with js_of_ocaml

````ocaml
let register_file fname = 
    match Iocaml.load_from_server fname with
    | Some(x) -> Sys_js.register_file ~name:fname ~content:x
    | None -> ()
````

Load the cmo and cmi file.  It's very simple and just contains

```ocaml
let a = 1
```

````ocaml
register_file "test/test.cmo";
register_file "test/test.cmi"
````

The run time needs a know where to find test.cmi

````ocaml
Topdirs.dir_directory "test"
````

Now turn on all the js_of_ocaml debugging options (Option.Debug.set "xxx" for parser,deadcode,main,linker,flow,times)

**actually we'll leave it off for this run, so we wont execute this cell**

````ocaml
Iocaml.jsoo_debug_on ()
````

Load the cmo file.  The first part of the trace is related to the actual code in the cell below.  The last 15 or so lines are where the bytecode file is being processed by js_of_ocaml.

````ocaml
Topdirs.dir_load Format.std_formatter "test/test.cmo"
````
